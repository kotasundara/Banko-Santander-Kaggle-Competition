## If you look at the above outputs, target feature is highly imbalanced. If I guess 0 without any model, nearly 90% of the time, I am correct. This is completely misleading. We will need to deal with this highly imbalanced data as otherwise machine learning algorithms are generally designed to predict what the majority observations say i.e. in this case 0. Performance metric in these cases is NOT the accuracy percentage but AUC under ROC is the best performance metric. In such imbalanced cases, ensemble methods generally outperform other ML models. Go for Decision Tree, random forest and gradient boost tree models.

```python
X=train.drop(columns=["ID_code","target"]).values # prepare X 
y=train["target"].values            # prepare y
```

```python
print(X.shape)
print(y.shape)
```

Highly Imbalanced are the target labels: Let us use SMOTE(synthetic minority oversampling technique)

```python
from imblearn.over_sampling import SMOTE

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

print("Number transactions X_train dataset: ", X_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", X_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
# split the training dataset into train and test datasets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier
#Create a Decision Tree classifier
model2 = DecisionTreeClassifier()

# Train the model using the training sets 
model2.fit(X_train,y_train)
pred=model2.predict(X_test)
```

# Use Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
#Create a Random Forest classifier
model3 = RandomForestClassifier()

# Train the model using the training sets 
model3.fit(X_train,y_train)
pred=model3.predict(X_test)
```

```python
# Random Forest accuracy score and auc scores
pred_prob=model3.predict_proba(X_test)
print("Random Forest. accuracy_score:",accuracy_score(y_test,pred))
print("Random Forest. auc score:",roc_auc_score(y_test,pred_prob[:,1]))
```
